## Question 1.2 (1 Point)
Copy **YOUR** implementation of ```StochasticGradientDescent``` from [Assignment 5](../assignment-5/assignment-5.ipynb) into the following cell.

````ruby
# BEGIN YOUR CODE
# raise NotImplementedError.new()

class StochasticGradientDescent
  attr_reader :weights
  attr_reader :objective
  def initialize obj, w_0, lr = 0.01
    @objective = obj
    @weights = w_0
    @n = 1.0
    @lr = lr
  end
  def update x
    # BEGIN YOUR CODE
    curr_lr = @lr / Math.sqrt(@n)
    @objective.func(x, @weights)
    grad = @objective.grad(x, @weights)
    @weights = update_weights(@weights, grad, curr_lr)
    @n += 1
    #END YOUR CODE
  end
  
  def update_weights(w, dw, lr)
    w_copy = w.clone()
    dw_copy = dw.clone()
  
    dw_copy.each do |k, v|
      dw_copy[k] *= lr
    end
  
    w_copy.each do |k, v|
      if dw_copy.key?(k)
        w_copy[k] -= dw_copy[k]
      end
    end
    w_copy
  end
  
end
#END YOUR CODE
````

## Question 1.3 (5 Points)


Refactor **your** implementation of the AUC metric to return the false positive, true positive values for all scores in addition to the AUC. We will use these to plot the ROC curve.

````ruby
def roc_curve(scores)
  # BEGIN YOUR CODE
  res = Array.new
  
  min = (scores.min{|a,b| a[0] <=> b[0]})[0]
  max = (scores.max{|a,b| a[0] <=> b[0]})[0]
  num = 1000.0
  tp = Array.new
  fp = Array.new
  step = (max - min) / num
  
  (min..max).step(step).to_a.each do |threshold|
    false_neg = 0.0 
    true_pos = 0.0
    true_neg = 0.0
    false_pos = 0.0
    scores.each do |record|
      if(record[1] == 0 or record[1] == -1) and record[0] >= threshold
        false_pos += 1.0
      end
      if(record[1] == 0 or record[1] == -1) and record[0] < threshold
        true_neg += 1.0
      end
      if(record[1] == 1 and record[0] >= threshold)
        true_pos += 1.0
      end
      if(record[1] == 1 and record[0] < threshold)
        false_neg += 1.0
      end
    end
    fp_temp = false_pos / (false_pos + true_neg)
    tp_temp = true_pos / (true_pos + false_neg)
    fp << fp_temp
    tp << tp_temp
    
    res << [fp_temp, tp_temp]    
  end
  auc = 0.0
  sorted = res.sort{|a,b| b[0] <=> a[0]}
  (1..sorted.length - 1).each do |index|
    width = sorted[index - 1][0] - sorted[index][0]
    height = sorted[index][1]
    auc += width * height
  end

  #END YOUR CODE
  return [fp, tp, auc]
end
````

## Question 2.2 (10 Points)

Implement the objective function, which is defined as follows for a mini-batch of $n$ examples:

# $L(w,X) = \frac{\lambda}{2} \left\lVert w \right\rVert ^ 2 + \frac{1}{n} \sum_{i} \max \left[ 0, 1 - y_i \widehat{y}_i\right]$

where ```@reg_param``` is the value $\lambda$.

Remember that for hinge loss, the label must be $y = \pm 1$.

````ruby
class HingeLossClassifier
  def func data, w
    # BEGIN YOUR CODE
#     raise NotImplementedError.new()
    res = 0.0
    size = data.length
    data.each do |record|
      label = record["label"] == 0 ? -1 : record["label"]
      predict_val = predict record, w
      update_val = ([0, 1 - label * predict_val].max)
      res += update_val
    end
    
    res /= size
    res += (@reg_param * norm(w) ** 2 / 2)
    #END YOUR CODE
    res
  end
end
````

## Question 2.3 (10 Points)

Implement the Hinge Loss classifier gradient method.


````ruby
class HingeLossClassifier
  def grad data, w
    g = Hash.new {|h,k| h[k] = 0.0}
    # BEGIN YOUR CODE
    size = data.length
    data.each do |record|
      label = record["label"] == 0 ? -1 : record["label"]
      predict_val = predict record, w
      record["features"].each do |key, value|
        if(!g.key?(key))
          g[key] = 0.0
        end
        if ((label * predict_val) < 1)
          g[key] -= label * value / size
        end
      end
    end
    
    g.each do |key, value|
      g[key] = value + @reg_param * w[key]
    end
    
    #END YOUR CODE
    return g
  end
end
````

## Question 3.1 (10 Points)

Implement the linear kernel where $x_i$ and $x_j$ are rows in a dataset.

````ruby
class LinearKernel
  def func x_i, x_j
    # BEGIN YOUR CODE
    res = 0.0
    
    features_a = x_i["features"]
    features_b = x_j["features"]
    
    features_a.each do |key, value|
      if features_b.key?(key)
        res += value * features_b[key]
      end
    end
    
    res
    #END YOUR CODE
  end
end
````

For any new datapoint z, the _Gram vector_ is defined as follows:

Note that on the training $g(x_i) = G_{i,\cdot}$.

````ruby
class GramMatrix
  def initialize kernel
    @kernel = kernel
    @data = Array.new
  end
  
  def train data
    # BEGIN YOUR CODE
    @data = data
    # END YOUR CODE
  end
  
  def transform z
    # BEGIN YOUR CODE
    res = Hash.new
    if z["label"] != nil
      res["label"] = z["label"]
    end
    res["features"] = Hash.new
    
    size = @data.length
    size.times do |i|
      key = "k_#{i}"
      res["features"][key] = @kernel.func(z, @data[i])
    end
    res
    # END YOUR CODE
  end
end
````
